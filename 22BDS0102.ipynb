```python
# Convert categorical variables
df_encoded = df.copy()
df_encoded['gender'] = df_encoded['gender'].map({'male': 1, 'female': 0})
df_encoded['married'] = df_encoded['married'].map({'yes': 1, 'no': 0})

# Drop rownames if present
df_encoded = df_encoded.drop(columns=['rownames'], errors='ignore')

# Normalization (since features are on different scales)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(df_encoded)
```

```python
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt

sc = SpectralClustering(n_clusters=3, affinity='nearest_neighbors', random_state=42)
y_spectral = sc.fit_predict(X)

plt.scatter(df['prestige'], df['mentor'], c=y_spectral, cmap='viridis')
plt.title("Spectral Clustering (Prestige vs Mentor)")
plt.xlabel("Prestige")
plt.ylabel("Mentor")
plt.show()
```

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(X)

plt.scatter(df['prestige'], df['mentor'], c=gmm_labels, cmap='coolwarm')
plt.title("Model-based Clustering (GMM using EM)")
plt.xlabel("Prestige")
plt.ylabel("Mentor")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch

# Dendrogram
plt.figure(figsize=(6,4))
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()

# Clustering
hc = AgglomerativeClustering(n_clusters=3, linkage='ward')
y_hc = hc.fit_predict(X)

plt.scatter(df['prestige'], df['mentor'], c=y_hc, cmap='Accent')
plt.title("Hierarchical Agglomerative Clustering")
plt.xlabel("Prestige")
plt.ylabel("Mentor")
plt.show()
```

```python
from scipy.spatial import distance_matrix
from scipy.sparse.csgraph import minimum_spanning_tree
import numpy as np

dist_matrix = distance_matrix(X, X)
mst = minimum_spanning_tree(dist_matrix).toarray()

print("Minimum Spanning Tree Adjacency Matrix:\n", np.round(mst, 2))
```

```python
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(n_neighbors=3)
outlier_labels = lof.fit_predict(X)

plt.scatter(df['prestige'], df['mentor'], c=outlier_labels, cmap='coolwarm')
plt.title("Outlier Detection (Local Outlier Factor)")
plt.xlabel("Prestige")
plt.ylabel("Mentor")
plt.show()

outliers = df[outlier_labels == -1]
print("\nOutliers detected:\n", outliers)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Copy and preprocess your dataset
df_encoded = df.copy()

# Convert categorical to numeric
df_encoded['gender'] = df_encoded['gender'].map({'male': 1, 'female': 0})
df_encoded['married'] = df_encoded['married'].map({'yes': 1, 'no': 0})

# Drop non-numeric columns if present
df_encoded = df_encoded.drop(columns=['rownames'], errors='ignore')

# Scale
scaler = StandardScaler()
X = scaler.fit_transform(df_encoded)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Explained variance
print("Explained Variance Ratio:", pca.explained_variance_ratio_)

plt.scatter(X_pca[:,0], X_pca[:,1], c='skyblue', s=80)
plt.title("PCA Projection (2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
import numpy as np

U, S, VT = np.linalg.svd(X)
print("Singular Values:", S)

# Show first two singular vectors
plt.plot(S, 'o-', color='purple')
plt.title("Singular Values (SVD)")
plt.xlabel("Component Index")
plt.ylabel("Singular Value")
plt.show()
```

```python
from sklearn.decomposition import FactorAnalysis

fa = FactorAnalysis(n_components=2, random_state=42)
X_fa = fa.fit_transform(X)

plt.scatter(X_fa[:,0], X_fa[:,1], c='teal', s=80)
plt.title("Factor Analysis Projection (2D)")
plt.xlabel("Factor 1")
plt.ylabel("Factor 2")
plt.show()
```

```python
explained_var = np.cumsum(pca.explained_variance_ratio_)
plt.plot(range(1, len(explained_var)+1), explained_var, 'o-', color='darkorange')
plt.axhline(0.95, color='red', linestyle='--')
plt.title("Intrinsic Dimensionality (Explained Variance)")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.show()
```

```python
from sklearn.manifold import MDS

mds = MDS(n_components=2, random_state=42, dissimilarity='euclidean')
X_mds = mds.fit_transform(X)

plt.scatter(X_mds[:,0], X_mds[:,1], c='lightcoral', s=80)
plt.title("Multidimensional Scaling (MDS)")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.show()
```

```python
from sklearn.manifold import Isomap

iso = Isomap(n_neighbors=4, n_components=2)
X_iso = iso.fit_transform(X)

plt.scatter(X_iso[:,0], X_iso[:,1], c='gold', s=80)
plt.title("Manifold Learning (Isomap)")
plt.xlabel("Manifold Dimension 1")
plt.ylabel("Manifold Dimension 2")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder

# Assuming df is already loaded
# Encode categorical variables
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
df['married'] = le.fit_transform(df['married'])

# Define features (X) and target (y)
X = df[['articles', 'gender', 'married', 'kids', 'mentor']]
y = df['prestige']

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Build model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("Linear Regression Model Evaluation")
print("R Score:", round(r2, 3))
print("MSE:", round(mse, 3))
print("RMSE:", round(rmse, 3))

# Display coefficients
coef_df = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
print("\nModel Coefficients:\n", coef_df)
```
